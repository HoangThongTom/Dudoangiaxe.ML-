```python
!python -m pip install plotly matplotlib seaborn --quiet
import plotly.express as px
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Thiết lập style mặc định cho biểu đồ
sns.set_style("darkgrid")
matplotlib.rcParams["font.size"] = 14
matplotlib.rcParams["figure.figsize"] = (10, 6)
matplotlib.rcParams["figure.facecolor"] = "#00000000"
```

```python
import pandas as pd
df_raw = pd.read_csv("Data_CarPrice.csv")
df_raw.describe()
```

```python
# Cột số: Age, kmDriven, AskPrice, km_per_year
numeric_cols = ["Age", "kmDriven", "AskPrice", "km_per_year"]

for col in numeric_cols:
    fig = px.histogram(df_raw, 
                       x=col, 
                       marginal="box",
                       title=f"Phân bố {col}")
    fig.update_layout(bargap=0.1)
    fig.show()
```

```python
# Cột loại: Brand, Transmission, Owner, FuelType
categorical_cols = ["Brand", "Transmission", "Owner", "FuelType"]

for col in categorical_cols:
    fig = px.bar(df_raw[col].value_counts().reset_index(),
                 x=col, 
                 y="count",
                 title=f"Phân bố {col}")
    fig.show()
```

```python
# Cột thời gian: post_month
fig = px.bar(df_raw["post_month"].value_counts().reset_index(),
             x="post_month", y="count",
             title="Số lượng xe đăng bán theo tháng")
fig.show()
```

```python
!python -m pip install statsmodels
# Age với AskPrice 
'''
- Xe càng cũ thì giá càng thấp (xu hướng âm)
- xe cùng tuổi nhưng vẫn có những mức giá chênh lệch 
'''
fig = px.scatter(df_raw,
                 x="Age",
                 y="AskPrice",
                 title="Age vs AskPrice",
                 trendline="ols")  # đường xu hướng
fig.show()
```

```python
# kmDriven với AskPrice: Quãng đường càng ngắn thì giá thành càng cao
df_raw["km_group"] = pd.cut(df_raw["kmDriven"], 
                             bins=[0, 20000, 40000, 60000, 80000, 100000, float("inf")],
                             labels=["0-20k", "20-40k", "40-60k", "60-80k", "80-100k", "100k+"])

fig = px.box(df_raw,
             x="km_group",
             y="AskPrice",
             title="AskPrice theo nhóm kmDriven",
             color="km_group")
fig.show()
```

```python
#FuelType vs giá
fig = px.bar(df_raw.groupby("FuelType")["AskPrice"].mean().reset_index(),
             x="FuelType",
             y="AskPrice",
             title="Giá trung bình theo FuelType",
             color="FuelType")       
fig.show()
```

```python
#Transmission vs giá
fig = px.bar(df_raw.groupby("Transmission")["AskPrice"].mean().reset_index(),
             x="Transmission",
             y="AskPrice",
             title="Giá trung bình theo Transmission",
             color="Transmission")
fig.show()
```

```python
# Owner với giá
fig = px.bar(df_raw.groupby("Owner")["AskPrice"].mean().reset_index(),
             x="Owner",
             y="AskPrice",
             title="Giá trung bình theo số chủ sở hữu (Owner)",
             color="Owner")
fig.show()
```
